```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
img_bgr = cv2.imread('Data/Lenna.png',1)
```

```python
img1 = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
```

```python
img_bgr = cv2.imread('Data/Interstellar.jpg',1)
```

```python
img2 = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
```

```python
def bilinear_interpolation(img, x, y):
    h, w, c = img.shape
    x0, y0 = int(np.floor(x)), int(np.floor(y))
    x1, y1 = min(x0 + 1, w - 1), min(y0 + 1, h - 1)

    a, b = x - x0, y - y0

    pixel = (1 - a) * (1 - b) * img[y0, x0] + a * (1 - b) * img[y0, x1] + \
            (1 - a) * b * img[y1, x0] + a * b * img[y1, x1]

    return pixel.astype(np.uint8)
```

```python
def affine_transformation(image , tx=0 ,ty=0 , theta=0 ,sx=1 , sy=1 , shx=0 , shy=0):

    theta=np.radians(theta)
    
    h, w, c = image.shape

    cx,cy= w//2 , h//2
    
    # 1. Translation Matrix
    T = np.array([
        [1, 0, -tx],
        [0, 1, -ty],
        [0, 0, 1]
    ])

    # 2. Rotation Matrix
    R = np.array([
        [np.cos(theta), -np.sin(theta), 0],
        [np.sin(theta),  np.cos(theta), 0],
        [0, 0, 1]
    ])

    # 3. Scaling Matrix
    S = np.array([
        [sx, 0, 0],
        [0, sy, 0],
        [0, 0, 1]
        ])

    # 4. Shearing Matrix in the X-direction
    Shx = np.array([
        [1, shx, 0],
        [0, 1, 0],
        [0, 0, 1]
    ])

    # 5. Shearing Matrix in the Y-direction
    Shy = np.array([
        [1, 0, 0],
        [shy, 1, 0],
        [0, 0, 1]
    ])

    # 6. Center translation matrices
    cen = np.array([
        [1, 0, -cx],
        [0, 1, -cy],
        [0, 0, 1]
    ])

    cen_inv = np.array([
        [1, 0, cx],
        [0, 1, cy],
        [0, 0, 1]
    ])

    transformation_matrix = T @ cen_inv @ R @ S @ Shx @ Shy @ cen
    
    transformed_image = np.zeros_like(image)
    for y in range(h):
        for x in range(w):
            new_coords = np.dot(transformation_matrix, np.array([x, y, 1]))
            new_x, new_y = new_coords[0], new_coords[1]

            if 0 <= new_x < w and 0 <= new_y < h:
                transformed_image[y, x] = bilinear_interpolation(image, new_x, new_y)

    return transformed_image
```

```python
fig , axes = plt.subplots(8,2,figsize=(10,20))

title = ['Original' , 'Translation (x by 30)' , 'Translation (y by 40)' , 'Rotation ( by 45)' , 'Scaling (x by 2)' , 'Scaling (y by 3) ' , 'Shearing (x by 1.2)' , 'Shearing (y by 0.6)']

img = [img1,img2]

lst = [0,30,40,45,2,3,1.2,0.6]

for i in range(8):

    params_dict = {
        'tx': 0 if i != 1 else lst[i],
        'ty': 0 if i != 2 else lst[i],
        'theta': 0 if i != 3 else lst[i],
        'sx': 1 if i != 4 else lst[i],
        'sy': 1 if i != 5 else lst[i],
        'shx': 0 if i != 6 else lst[i],
        'shy': 0 if i != 7 else lst[i],
        }
    for j in [0,1]:
        new_img=affine_transformation(img[j] , **params_dict)
        axes[i , j].imshow(new_img)
        axes[i,j].set_title(title[i])
        axes[i,j].axis('off')
```

```python
fig , axes = plt.subplots(4,2,figsize=(8,10))

title = ['-x y' , 'x y' , '-x -y' ,'x -y']

lst = [[-1,1] , [1,1] , [-1,-1] , [ 1,-1]]

for i in range(2):
    for j in range(4):
        a,b = lst[j]
        new_img=affine_transformation(img[i] , sx = a, sy=b)
        n=i*4 +j
        axes[n//2 , n%2].imshow(new_img)
        axes[n//2 , n%2].set_title(title[n%4])
        axes[n//2 , n%2].axis('off')
```
